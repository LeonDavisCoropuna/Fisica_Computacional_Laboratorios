```python
import numpy as np
import matplotlib.pyplot as plt

def rule_selection(left, current, right):
    """Aplica la regla de selección para determinar el nuevo estado de la celda actual."""
    if (left, current, right) == (0, 0, 0):
        return 0
    elif (left, current, right) == (0, 0, 1):
        return 1
    elif (left, current, right) == (0, 1, 0):
        return 1
    elif (left, current, right) == (0, 1, 1):
        return 1
    elif (left, current, right) == (1, 0, 0):
        return 1
    elif (left, current, right) == (1, 0, 1):
        return 0
    elif (left, current, right) == (1, 1, 0):
        return 0
    elif (left, current, right) == (1, 1, 1):
        return 0


def generate_next_state(current_state):
    """Genera el siguiente estado del autómata celular."""
    next_state = np.zeros_like(current_state)
    for i in range(1, len(current_state) - 1):
        left = current_state[i - 1]
        current = current_state[i]
        right = current_state[i + 1]
        next_state[i] = rule_selection(left, current, right)
    return next_state


def cellular_automaton(initial_state, steps):
    """Genera un autómata celular y devuelve la evolución a lo largo de varios pasos."""
    states = [initial_state]
    current_state = initial_state
    for _ in range(steps):
        next_state = generate_next_state(current_state)
        states.append(next_state)
        current_state = next_state
    return np.array(states)

# Parámetros iniciales
size = 200  # Tamaño del autómata
steps = 100  # Número de pasos
initial_state = np.zeros(size, dtype=int)
initial_state[size // 2] = 1  # Iniciar con una única celda activada en el centro

# Generar el autómata celular
states = cellular_automaton(initial_state, steps)

# Visualizar el autómata celular
plt.figure(figsize=(10, 10))
plt.imshow(states, cmap="binary", interpolation="nearest")
plt.title("Autómata Celular")
plt.xlabel("Celdas")
plt.ylabel("Pasos")
plt.show()
```
